Una organización requiere de datos, y sobre todo del análisis de estos, para tomar mejores decisiones. Sin embargo, para que los datos aporten valor, es importante que cumplan con ciertas características como la accesibilidad, disponibilidad, facilidad de lectura y consulta, calidad o precisión. Es aquí donde entran los conceptos de almacenamiento de datos y bases de datos.
	Una base de datos es una colección de piezas de información organizadas y almacenadas en un repositorio que permite que las diversas partes interesadas de una organización puedan consultar esa información de forma fácil y rápida. Los tipos de bases de datos más comunes son las bases de datos relacionales y las bases de datos no relacionales; una forma sencilla de imaginar las bases de datos relacionales es pensar en diversas tablas, cada una con diferente información, pero, que tienen la capacidad de “relacionarse” entre sí por medio de agregación de datos entre una tabla y otra, gracias a una columna o tipo de información en común. Para acceder a las bases de datos relacionales, se utiliza el lenguaje de consultas llamado Structured Query Language (SQL). Mientras que las bases de datos no relacionales (también conocidas como bases de datos No-SQL) almacenan su información por medio de categorías, en vez de tablas con relaciones; es decir, utilizando pares claves-valor (key-value). Una manera de imaginar estas bases de datos sería pensar en un conjunto de diccionarios de Python, o de objetos JSON.
	Por último, es importante mencionar el concepto de almacenamiento de datos (data warehouse), el cual podemos visualizar como un single source of truth de una organización. A diferencia de las bases de datos que pudieran contener datos no procesados, un data warehouse funciona como el centro de datos más importante de una organización, es decir, esa ubicación central donde las partes interesadas tienen acceso de forma rápida y certera a los datos de diversas fuentes de información, pero datos que han sido detalladamente procesados y almacenados para su futuro análisis. Para ingresar datos a un data warehouse, los científicos e ingenieros de datos suelen realizar un proceso conocido como ETL (extract, transform and load), el cual se refiere a que primero se extraen los datos de una fuente, después se transforman para prepararlos de la manera óptima para su análisis, y por último se cargan a la base de datos correspondiente en el data warehouse (aunque, gracias al desarrollo de técnicas de big data, ahora también se utiliza el proceso de ELT –extract, load and transform–).

```python
#declarar dataframe con info de base de datos
tpd6=pd.DataFrame(data=taiwan_payments_data)
tpd6
```

```python
#promedio de las edades
am=tpd6.Age.mean()
am
```

```python
#reemplazar NaN de edades con promedio de las edades
tpd6['Age'].fillna(value=am,inplace=True)
tpd6
```

```python
#reemplazar faltantes de Gender con la media de Gender
tpd6['Gender'].fillna(value = tpd6.Gender.median(), 
                    inplace = True)
tpd6
```

```python
#el valor que aparece mas en la columna education
mm  = tpd6.Education.mode()
mm
```

```python
#rellenar espacios vacíos o indefinidos en education, con el valor que mas se repite en education
tpd6['Education'].fillna(value = mm, 
                    inplace = True)
tpd6
```
